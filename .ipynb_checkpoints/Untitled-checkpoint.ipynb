# Physicochemical Analysis & Predictive Modeling of Viticulture Quality

# Section-I: Defining the Problem Statement
## 1. The Context
In the wine industry, quality assessment is traditionally a subjective process involving human "sommeliers." However, sensory perception is inconsistent and expensive. There is a significant industrial need to correlate **objective physicochemical properties** (like pH, residual sugar, and alcohol content) with **perceived quality scores** to automate grading and ensure production consistency.

## 2. The Problem Definition
The goal of this project is to develop a robust mathematical framework to model the relationship between 11 chemical input features and a discrete quality rating (3–8) using a dataset of approximately 1,600 red wine samples.
### The core challenge is two-fold:
a) **Predictive Accuracy:** Determining whether a wine sample is "high quality" based on chemical signatures.

b) **Feature Significance:** Mathematically identifying which chemical markers (e.g., volatile acidity vs. alcohol) have the highest impact on the final quality score.

## 3. Technical Constraint
To ensure a deep understanding of the underlying algorithmic logic, this project implements ML algorithms like Linear Regression, Decision Tree, Random Foreest from scratch.

## 4. Success Criteria
The project will be considered successful if the "from-scratch" implementation achieves:

a) **Parity:** Result (weight/bias) should align with established library benchmarks (Scikit-Learn) within a 0.01% margin of error.

b) **Interpretability:** A clear ranking of feature importance that aligns with viticulture logic (e.g., high volatile acidity usually leads to lower quality).

```python
import numpy as np

def stratified_train_test_split(dataset, test_ratio, target_col='quality'):
    train_indices = []
    test_indices = []
    
    # Loop through each unique quality grade (3, 4, 5, 6, 7, 8)
    for grade in dataset[target_col].unique():
        # Get indices for just this grade
        grade_indices = dataset[dataset[target_col] == grade].index.values.copy()
        
        # Shuffle the indices for this grade only
        np.random.seed(42)
        np.random.shuffle(grade_indices)
        
        # Calculate split point for this specific grade
        test_size = int(len(grade_indices) * test_ratio)
        
        # Distribute the indices
        test_indices.extend(grade_indices[:test_size])
        train_indices.extend(grade_indices[test_size:])
        
    return dataset.loc[train_indices], dataset.loc[test_indices]

train_set, test_set = stratified_train_test_split(wine_data, 0.2)
```

## Observation: Limitations of Central Tendency via Bar Plots
While the initial bar plots suggest a linear relationship for **Citric Acid** and **Alcohol**, the visualizations for the remaining physicochemical features (notably Residual Sugar, Chlorides, and Sulfur Dioxide) appear statistically misleading.

## Technical Justification for Alternative Analysis:
a) **Skewness and Heavy Tails:** Features like Chlorides exhibit significant right-skewness. In these cases, the "Mean" represented by the bar height is heavily influenced by extreme values (outliers), failing to represent the typical profile of the wine sample.

b) **Information Loss:** Bar plots squash the distribution into a single point of central tendency. They hide the **variance** and **modality** (e.g., whether a feature has two distinct "peaks" of common values), which is critical for the clsssification logic.

c) **The Case for Transformation:** The presence of high-magnitude outliers in Residual Sugar suggests that the data may be better analyzed in a **Logarithmic Space**. Applying $x_{log} = \log(1 + x)$ will normalize the variance and prevent these extreme values from dominating the Gradient Descent optimization in later stages.

d) **Requirement for Density Estimation:** To better understand the feature-target relationship, we will pivot to **Violin Plots** or **Kernel Density Estimate (KDE) plots**. These will allow us to visualize the probability density of each feature across different quality grades, revealing whether "outliers" are random noise or distinct sub-varieties of wine.

### Fisher-Pearson Skewness
To calculate skewness, we look at the **third standardized moment**.

The formula for a population skewness is: $$\gamma = \frac{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^3}{\left(\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2\right)^{3/2}}$$

**Numerator:** The average of the cubed deviations from the mean (emphasizes the tail)

**Denominator:** The standard deviation cubed (normalizes the scale)

| Skewness Score | Interpretation |
|:--------------:|:--------------:|
| >1.0           | Highly-Slewed  |
| 0.5 to 1.0     | Moderately Skewed |
| -0.5 to 0.5    | Approximately Symmetric |

```python
def skewness_score(feature_data):
    n = len(feature_data)
    mean = np.mean(feature_data)

    # numerator calculation (3rd moment)
    numerator = np.sum((feature_data - mean) ** 3) / n
    # denominator calculation (standard deviation cubed)
    denominator = (np.sum((feature_data - mean) ** 2) / n) ** (1.5)

    return numerator / denominator

skew_report = {}
for col in feature_list:
    skew_report[col] = skewness_score(train_set[col])

for feature, score in sorted(skew_report.items(), key=lambda x: x[1], reverse=True):
    print(f"{feature:20} : {score:.4f}")
```

We are now gonna use: 

a) log transform for highly-skewed features

b) robust scaler for moderately-skewed features

c) standard scaler (Z-score) for the rest

$$f(x) = 
\begin{cases} 
\text{Standardize}(\ln(x+e)) & \text{if } \gamma > 1.0 \\
\text{RobustScale}(x) & \text{if } 0.5 < \gamma \leq 1.0 \\
\text{Standardize}(x) & \text{if } \gamma \leq 0.5 
\end{cases}$$
where $\gamma$ is the Fisher-Pearson Skewness coefficient.

Formula for Standard Scaler: $$x_{scaled} = \frac{x - \text{Mean}(x)}{\sigma}$$

```python
def standard_scaler(feature_data, mean, std):
    n=len(feature_data)
    # calculation of standard deviation
    # std=(np.sum((feature_data - mean) ** 2) / n) ** (0.5)
    
    # Adding a small epsilon to std avoids division by zero
    return (feature_data-mean)/(std+1e-7)
```

Formula for Robust scaler: $$x_{scaled} = \frac{x - \text{Median}(x)}{Q3(x) - Q1(x)}$$

```python
def robust_scaler(feature_data, median, iqr):
    median=np.median(feature_data)
    
    # Adding a tiny epsilon to IQR avoids division by zero
    return (feature_data-median)/(iqr+1e-7)
```

Foemula for log transform: $$x_{log} = \ln(1 + x)$$

1+x ensures that the result is always defined for non-negative values.

The log(1+x) method doesn't work well on reducing skewness of "chlorides","residual sugar", "sulphates" because the values are very tiny )between 0.01 to 0.1. To fix this we are gonna use $x_{log} = \log(x + e)$ for them.

```python
def robust_params(feature_data):
    median=np.median(feature_data)
    q1=np.percentile(feature_data,25)
    q3=np.percentile(feature_data,75)
    iqr=q3-q1
    return median, iqr
```

We use the same scaling parameters derived from train_set for scaling of both train and test dataset.

```python
processed_train_set=train_set.copy()
processed_test_set=test_set.copy()

for feature, score in skew_report.items():
    raw_values_train=train_set[feature]
    raw_values_test=test_set[feature]
    if score > 1.0:
        # To check if the values are too small
        if np.mean(raw_values_train) < 1.0:
            log_data_train=np.log(raw_values_train + 1e-7)
            log_data_test=np.log(raw_values_test + 1e-7)
        else:
            log_data_train=np.log1p(raw_values_train)
            log_data_test=np.log1p(raw_values_test)
        mean=np.mean(log_data_train)
        std=np.std(log_data_train)
        processed_train_set[feature]=standard_scaler(log_data_train, mean, std)
        processed_test_set[feature]=standard_scaler(log_data_test, mean, std)
    elif score > 0.5:
        median, iqr=robust_params(raw_values_train)
        processed_train_set[feature]=robust_scaler(raw_values_train, median, iqr)
        processed_test_set[feature]=robust_scaler(raw_values_test, median, iqr)
    else:
        mean=np.mean(raw_values_train)
        std=np.std(raw_values_train)
        processed_train_set[feature]=standard_scaler(raw_values_train, mean, std)
        processed_test_set[feature]=standard_scaler(raw_values_test, mean, std)

display(processed_train_set.head())
```

## Section-IV: Building Regression Models
## Model-1: Simple Linear Regression
### Method-A: Mean Square Error Optimisation Uisng Gradient Descent
**Mean Square Error Cost Function:** $$J(W) = \frac{1}{2m} \sum_{i=1}^{m} (h_w(x^{(i)}) - y^{(i)})^2$$
where $h_w = w^{T}x$

The **Gradient** of MSE is given by: $$\nabla J(W) = \frac{1}{m} \sum_{i=1}^{m} (h_w(x^{(i)}) - y^{(i)}) \cdot x^{(i)}$$

We can translate this summation above into a much faster **Vectorized Matrix Multiplication** which avoids using a for loop for that $\sum$ symbol and calculate the entire gradient for all features at once: $$\nabla J(W) = \frac{1}{m} X^T (XW - y)$$
where $XW - y$ is our "Error Vector".

```python
# Extracting features and label from the processed_train_set.
X_train=processed_train_set.drop('quality', axis=1)
y_train=processed_train_set['quality'].copy()

# Extracting features and label form processed_test_set.
X_test=processed_test_set.drop('quality', axis=1)
y_test=processed_test_set['quality'].copy()

# Adding a column of 1s to the features matrix.
# Which represents the 'intercept' (w_0)
intercept_ones_train=np.ones((X_train.shape[0], 1))
X_train_final=np.hstack((intercept_ones_train, X_train))
intercept_ones_test=np.ones((X_test.shape[0], 1))
X_test_final=np.hstack((intercept_ones_test, X_test))
```

### Linear Regression Using Gradient Descenet

```python
class LinearRegressionGD:
    def __init__(self, learning_rate, iterations):
        self.learning_rate=learning_rate
        self.iterations=iterations
        self.weights = None
        self.cost_history=[] # This is for checking that the model is learning
        
    def fit(self, X, y):
        # m= number of samples, n= number of features (including bias)
        m, n = X.shape
        
        # Ensurisng y is a true column vector (m, 1), for matrix math safety
        if hasattr(y, "values"): #It strip away the Pandas "wrapper" and keeps the raw numbers
            y=y.values.reshape(-1,1)
        else:
            y=y.reshape(-1,1)
        
        # Initilizing the weights to zeros
        self.weights=np.zeros((n,1))
        
        # The Gradient Descent Loop
        for i in range(self.iterations):
            predictions=np.dot(X, self.weights)
            error=predictions-y
            gradients=(1/m)*np.dot(X.T, error)
            
            # Updating weights
            self.weights -= self.learning_rate*gradients
            
            # Recording the cost (MSE) at each iteration
            cost=(1/m)*np.sum(error**2)
            self.cost_history.append(cost)

            # Displaying progress at every 100 steps
            if i % 100 == 0:
                print(f"Iteration {i} | Cost (MSE): {cost:.4f}")
    
    def predict(self, X):
        if self.weights is None:
            raise ValueError("The model is not fitted")
        if X.shape[1] != self.weights.shape[0]:
            raise ValueError(f"Feature mismatch! Model expects {self.weights.shape[0]} features, but got {X.shape[1]}.")
        
        return np.dot(X, self.weights)
    
# Initialization of the model
lr_model1 = LinearRegressionGD(0.1, 1000)

# Training the model
lr_model1.fit(X_train_final, y_train)

#Plotting the learning curve
plt.figure(figsize=(8, 5))
plt.plot(lr_model1.cost_history, color="blue", linewidth=2)
plt.title("Gradient Descent: Cost vs. Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.show()
```

```python
# Printing the final weights after training
feature_names= ['intercept']+list(feature_list)
if lr_model1.weights is not None:
    for name, weight in zip(feature_names, lr_model1.weights.flatten()):
        print(f"{name:20} : {weight:10.4f}")
```

```python
# Predicting the test dataset
y_test_pred = lr_model1.predict(X_test_final)
test_mse = np.mean((y_test_pred - np.asarray(y_test).reshape(-1,1))**2)
print(f"Test MSE: {test_mse:.4f}")
```

### Method-B: Closed form solution
Gradient Descent gives one possible mean for minimizing, which uses iterative approach and may take time. In the situation where we know that our cost function is strictly concave or convex, we can explicitly take its derivative to zero. This process of such derivation is called obtaining the normal equations or closed form.
Formula of closed form solution for the above MSE cost function is: $$W = (X^T X)^{-1} X^T y$$
We can use closed form solution if it exists and our dataset doesn't have too many features (otherwise the inverse computation will be slow).

```python
from numpy.linalg import inv

class LinearRegressionCF:
    def __init__(self):
        self.weights=None
        self.cost=0
    
    def fit(self, X, y): 
        m, n = X.shape       
        # Ensurisng y is a true column vector (m, 1), for matrix math safety
        if hasattr(y, "values"): #It strip away the Pandas "wrapper" and keeps the raw numbers
            y=y.values.reshape(-1,1)
        else:
            y=y.reshape(-1,1)
        
        # closed form solution
        self.weights=inv(X.T@X)@X.T@y
        
        #calculating MSE
        predictions=np.dot(X, self.weights)
        error=predictions-y
        self.cost=(1/m)*np.sum(error**2)
        print(f"Train MSE: {self.cost:.4f}")
    
    def predict(self, X):
        if self.weights is None:
            raise ValueError("The model is not fitted")
        if X.shape[1] != self.weights.shape[0]:
            raise ValueError(f"Feature mismatch! Model expects {self.weights.shape[0]} features, but got {X.shape[1]}.")
        
        return np.dot(X, self.weights)

# Initialization of the model
lr_model2=LinearRegressionCF()
# Trainig the model
lr_model2.fit(X_train_final, y_train)

# Displaying the weights after training
if lr_model2.weights is not None:
    for name, weight in zip(feature_names, lr_model2.weights.flatten()):
        print(f"{name:20} : {weight:10.4f}")
```

```python
# Predicting the test dataset
y_test_pred = lr_model2.predict(X_test_final)
test_mse = np.mean((y_test_pred - np.asarray(y_test).reshape(-1,1))**2)
print(f"Test MSE: {test_mse:.4f}")
```

## Model-2: Decision Tree

### The Limitation of Global Linearity
Linear Regression operates under the "Global Linearity" assumption: that each chemical feature (like Alcohol or Acidity) has a consistent, additive effect on quality across the entire spectrum.

In oenology (the study of wine), chemistry is rarely linear. For example:

a) **Feature Interation:** Volatile acidity might be tolerable in a high-alcohol wine but ruinous in a low-alcohol one. Linear models cannot "see" this interaction without manually created polynomial terms.

b) **Threshold Effects:** Certain chemicals may have no impact until they reach a specific "tipping point," at which quality drops or rises sharply.

##  The Case for Decision Tree
A Decision Tree Regressor replaces the linear equation with a series of hierarchical "if-then" splits. This allows the model to partition the feature space into localized regions.

a) **Non Linearity:** It naturally captures non-linear relationships by segmenting data based on variance reduction (MSE).

b) **Automatic Interaction Detecetion:** For example, by splitting on "Alcohol" and then immediately splitting that branch on "Sulphates," the tree implicitly learns how these variables interact without human intervention.

c) **Mathematical Basis (MSE Reduction):** At each node, the tree chooses a split $s$ for feature $j$ that minimizes the sum of squared errors:$$\sum_{i: x_i \in R_1(j,s)} (y_i - \hat{y}_{R_1})^2 + \sum_{i: x_i \in R_2(j,s)} (y_i - \hat{y}_{R_2})^2$$
where, 

$\hat{y}_{R_1}$: The average wine quality of all samples in the first bucket.

$\hat{y}_{R_2}$: The average wine quality of all samples in the second bucket.

### Method-A: Decision Tree Regressor From Scratch

```python
# Each node either holds a "split" (feature and threshold) or 
# a "prediction" (the average value if it's a leaf).
class Node:
    def __init__(
        self, feature=None, threshold=None, left=None, right=None, *, value=None
    ):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value

    def is_leaf(self):
        return self.value is not None
    
# The core logic: at every node, find the split that drops the Residual Sum of Squared Errors (RSS)
# the most, then split the data and do it again.

class ScratchDecisionTreeRegressor:
    def __init__(self, max_depth=10, min_samples_split=2):
        # The min_sample_split is a stop condition to prevent the tree from creating a leaf for 
        # every single indivisual wine
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split 
        self.root = None

    def fit(self, X, y):
        self.X=np.asarray(X)
        self.y=np.asarray(y)
        # Ensure y is a flat array for easier mean calculations
        self.root = self._grow_tree(self.X, self.y.flatten())

    def _grow_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape

        # Base Case: Stop if we hit max depth or don't have enough samples
        if (
            depth >= self.max_depth
            or n_samples < self.min_samples_split
            or len(np.unique(y)) == 1 # Pure node: All wines in the current set have same quality
        ):
            return Node(value=np.mean(y))

        # Find the best split
        best_feature, best_thresh = self._best_split(X, y, n_features)

        # If no split improves RSS, make it a leaf
        if best_feature is None:
            return Node(value=np.mean(y))

        # Recursive step
        left_idxs = X[:, best_feature] <= best_thresh
        right_idxs = ~left_idxs

        left = self._grow_tree(X[left_idxs, :], y[left_idxs], depth + 1)
        right = self._grow_tree(X[right_idxs, :], y[right_idxs], depth + 1)

        return Node(feature=best_feature, threshold=best_thresh, left=left, right=right)

    def _best_split(self, X, y, n_features):
        best_rss = float("inf")
        split_idx, split_thresh = None, None

        for feat_idx in range(n_features):
            X_column = X[:, feat_idx]
            thresholds = np.unique(
                X_column
            )  # Check every unique value as a potential split

            for threshold in thresholds:
                # Calculate RSS for this specific split
                left_y = y[X_column <= threshold]
                right_y = y[X_column > threshold]

                if len(left_y) == 0 or len(right_y) == 0:
                    continue

                # The RSS we discussed: Sum of Squares in both buckets
                rss = np.sum((left_y - np.mean(left_y)) ** 2) + np.sum(
                    (right_y - np.mean(right_y)) ** 2
                )

                if rss < best_rss:
                    best_rss = rss
                    split_idx = feat_idx
                    split_thresh = threshold

        return split_idx, split_thresh
    
    def predict(self, X):
        self.X=np.asarray(X)
        return np.array([self._traverse_tree(x, self.root) for x in self.X])

    def _traverse_tree(self, x, node):
        if node.is_leaf():
            return node.value

        if x[node.feature] <= node.threshold:
            return self._traverse_tree(x, node.left)
        return self._traverse_tree(x, node.right)
```

Scaling the features doesn't impact the outcome of a decision tree model, so we can use either scaled features or raw features and the outcome will be same.

```python
decision_tree_model1=ScratchDecisionTreeRegressor(max_depth=4, min_samples_split=50)
decision_tree_model1.fit(X_train, y_train)

# Making Prediction
y_pred = decision_tree_model1.predict(X_train)
train_mse= mean_squared_error(y_train, y_pred)
print(f"Train MSE: {train_mse:.4f}")
```

### The Greedy Nature of Decision Tree
Decision Tree is a Greedy Algorithm.

It make the best possible split right now at the current node.

It doesn't look ahead to see if a slightly "worse" split now might lead to a much better split two steps down the line.

This is why a single tree can sometimes get stuck in a "local" pattern that doesn't generalize well (which probably what happened earlier to give worse MSE than fo linear regression), which leads right back to the **Random Forest** solution—where we use many greedy trees to balance each other out.

## The Case for Random Forest
While a single Decision Tree is powerful, it is notoriously "greedy" and prone to overfitting (high variance). A Random Forest solves this by using an ensemble technique called Bagging (Bootstrap Aggregating).

a) **Average of All Trees:** It trains hundreds of diverse trees on different subsets of the data and features, then averages their predictions. This cancels out the "noise" from individual trees.

b) **Feature Importance:** Random Forests provide a clear ranking of which chemical features (e.g., Alcohol vs. Chlorides) actually drive the variance in quality.

c) **Out-of-Bag (OOB) Error:** It provides a built-in validation mechanism, allowing us to estimate test performance during the training phase.



### Method-A: Random Forest from Scratch

```python
class ChildDTR(ScratchDecisionTreeRegressor):
    def __init__(self, max_depth=10, min_samples_split=2, max_features=None):
        super().__init__(max_depth, min_samples_split)
        self.max_features = max_features
        self.feature_importances = None

    def fit(self, X, y):
        X = np.asarray(X)
        y = np.asarray(y).flatten()
        # Initialize the importance array based on number of columns in X
        self.feature_importances_ = np.zeros(X.shape[1])
        self.root = self._grow_tree(X, y)

    def _best_split(self, X, y, n_features):
        best_rss_reduction = -1  # We want to maximize the reduction
        split_idx, split_thresh = None, None

        # Calculate parent RSS once to measure the "Gain" later
        parent_rss = np.sum((y - np.mean(y)) ** 2)

        # FEATURE RANDOMNESS: Select a random subset of feature indices
        # This is the "max_features" part required for Random Forest
        all_indices = np.arange(n_features)
        if self.max_features is not None and self.max_features < n_features:
            indices = np.random.choice(all_indices, self.max_features, replace=True)
        else:
            indices = all_indices

        for feat_idx in indices:
            X_column = X[:, feat_idx]
            thresholds = np.unique(X_column)

            for threshold in thresholds:
                left_y = y[X_column <= threshold]
                right_y = y[X_column > threshold]

                if len(left_y) == 0 or len(right_y) == 0:
                    continue

                # Calculate RSS of the two new "buckets"
                rss_left = np.sum((left_y - np.mean(left_y)) ** 2)
                rss_right = np.sum((right_y - np.mean(right_y)) ** 2)
                current_rss = rss_left + rss_right

                # Calculate the reduction (Gain)
                reduction = parent_rss - current_rss

                if reduction > best_rss_reduction:
                    best_rss_reduction = reduction
                    split_idx = feat_idx
                    split_thresh = threshold

        # Record the importance gain for the chosen feature before returning
        if split_idx is not None:
            self.feature_importances_[split_idx] += best_rss_reduction  

        return split_idx, split_thresh

class ScratchRandomForest:
    def __init__(self, n_trees=100, max_depth=4, min_samples_split=50, max_features="sqrt"):
        self.n_trees = n_trees
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.max_features = max_features
        
        self.trees = []
        self.oob_error_ = None
        self.feature_importances_ = None

    def fit(self, X, y):
        self.X = np.asarray(X)
        self.y = np.asarray(y).flatten()
        n_samples, n_features = self.X.shape
        
        # Feature Importance Tracker
        # We accumulate the variance reduction (RSS drop) for each feature across all trees
        importance_dict = np.zeros(n_features)
        
        # Out-Of-Bag (OOB) Trackers
        oob_predictions = np.zeros(n_samples)
        oob_counts = np.zeros(n_samples)

        for _ in range(self.n_trees):
            # --- BAGGING (Bootstrap Aggregating) ---
            # Randomly select row indices with replacement
            boot_indices = np.random.choice(n_samples, size=n_samples, replace=True)
            X_boot = self.X[boot_indices]
            y_boot = self.y[boot_indices]
            
            # --- OUT-OF-BAG INDICES ---
            # Find the rows that were left out of this specific bootstrap sample
            oob_indices = np.setdiff1d(np.arange(n_samples), boot_indices)
            
            # --- FEATURE RANDOMNESS ---
            # Determine how many features to look at per split
            if self.max_features == "sqrt":
                n_feats_to_use = int(np.sqrt(n_features))
            else:
                n_feats_to_use = n_features
                
            # Initializing the tree
            tree = ChildDTR(
                max_depth=self.max_depth, 
                min_samples_split=self.min_samples_split,
                max_features=n_feats_to_use 
            )
            
            # Train the tree
            tree.fit(X_boot, y_boot)
            self.trees.append(tree)

            # --- OOB PREDICTION AGGREGATION ---
            # Have the tree predict only the data it didn't see during training
            if len(oob_indices) > 0:
                X_oob = self.X[oob_indices]
                preds = tree.predict(X_oob)
                
                # Accumulate predictions to average them later
                oob_predictions[oob_indices] += preds
                oob_counts[oob_indices] += 1
                
            # --- FEATURE IMPORTANCE AGGREGATION ---
            if hasattr(tree, 'feature_importances_'):
                importance_dict += tree.feature_importances_

        # --- FINALIZE OOB ERROR ---
        # Only calculate MSE for samples that were left out at least once
        valid_oob = oob_counts > 0
        final_oob_preds = oob_predictions[valid_oob] / oob_counts[valid_oob]
        true_oob_y = self.y[valid_oob]
        
        self.oob_error_ = np.mean((true_oob_y - final_oob_preds) ** 2)
        
        # --- FINALIZE FEATURE IMPORTANCE ---
        # Normalize so all importances sum to 1.0 (100%)
        if np.sum(importance_dict) > 0:
            self.feature_importances_ = importance_dict / np.sum(importance_dict)

    def predict(self, X):
        X = np.asarray(X)
        
        # --- AVERAGE OF ALL TREES ---
        # Get predictions from every single tree
        tree_preds = np.array([tree.predict(X) for tree in self.trees])
        
        # Average the predictions along the columns (axis=0)
        return np.mean(tree_preds, axis=0)
```

```python
random_forest_model1=ScratchRandomForest(max_depth=10, min_samples_split=10)
random_forest_model1.fit(X_train, y_train)

y_pred=random_forest_model1.predict(X_train)
train_mse=mean_squared_error(y_train, y_pred)
print(f"Train MSE: {train_mse:.4f}")
print(f"oob error: {random_forest_model1.oob_error_:.4f}")
if random_forest_model1.feature_importances_ is not None:
    for name, weight in sorted(zip(feature_list, random_forest_model1.feature_importances_.flatten()), key=lambda x:x[1], reverse=True):
        print(f"{name:20} : {weight:10.4f}")
```

### Random Forest Model Fine-Tuning using Hyperparameter Optimisation

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

cv_shuffle = KFold(n_splits=5, shuffle=True, random_state=42)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 20, 50],
    'max_features': ['sqrt', 'log2', None]
}

grid_search = GridSearchCV(
    RandomForestRegressor(oob_score=True), 
    param_grid, 
    cv=cv_shuffle, 
    scoring='neg_mean_squared_error',
    n_jobs=-1 # Uses all the CPU cores!
)

grid_search.fit(X_train, y_train)

pd.set_option('display.max_colwidth', None)

results_df = pd.DataFrame(grid_search.cv_results_)
# Sort by rank to see the top 5 models immediately
top_models = results_df.sort_values(by='rank_test_score').head(5)

# To see only the best MSE and params:
display(top_models[['mean_test_score', 'params']])

print(f"Best Params: {grid_search.best_params_}")
print(f"Best CV MSE: {-grid_search.best_score_:.4f}")
```

## Defining the "Bayes Error" Ceiling
Achieving a Test MSE of 0.3675 while oob error is 0.3187 and Train MSE 0.0438 with the best fine-tuning parameters indicate that the model has fully exhausted the details available in the 11 chemical features. Any remaining error is classified as Irreducible Error ($E_{Bayes}$), likely stemming from:

a) **Subjective Stochasticity:** The target variable (Quality) is a human-assigned integer (3–8). Sensory evaluation is inherently noisy; two different sommeliers might rate the same bottle differently based on external factors.

b) **Unobserved Features:** Crucial variables such as grape variety, vineyard altitude, soil composition, and fermentation temperature are missing from the dataset.

c) **Measurement Precision:** The chemical properties (e.g., pH, sulphates) are recorded as floats, but any minute inaccuracy in the lab equipment introduces noise that no algorithm can resolve.

## Feature Importance
The Random Forest successfully mapped the complex chemistry of wine quality. While the linear model looked for "slopes," the random forest discovered "thresholds."

a) **Alcohol & Sulphates:** These remained the dominant predictors, acting as the primary drivers of "body" and structural integrity.

b) **Volatile Acidity:** Identified as the critical "negative" indicator. The forest correctly prioritized this as a non-linear gatekeeper; high vinegar content overrides all other positive chemical attributes.

c) **Interaction Effects:** By allowing a depth of 20, the model captured conditional relationships—for example, the impact of Citric Acid may change significantly depending on the Fixed Acidity levels.
